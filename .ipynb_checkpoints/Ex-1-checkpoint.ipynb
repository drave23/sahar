# Exercise 1.3 - Genres

```python
def DisplayGenreCount(dataset):
    genres = ['unknown','Action','Adventure','Animation','Childrens','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror',
    'Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
    genres_count = {}
    for genre in genres:
        count = 0
        genre_movies = dataset[genre] == 1
        genres_count[genre] = dataset[genre_movies].shape[0]
    
    #sorting the genres 
    genres_count_list = sorted(genres_count.items(), key=operator.itemgetter(1),reverse=True)
    # lets display the same result in the histogram
    fig = plt.figure(1, figsize=(18,13))
    ax2 = fig.add_subplot(2,1,2)
    y_axis = [genre[1] for genre in genres_count_list]
    x_axis = [i for i in range(len(genres_count))]
    x_label = [genre[0] for genre in genres_count_list]
    plt.xticks(rotation=85, fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.xticks(x_axis, x_label)
    plt.ylabel("No. of occurences", fontsize = 24, labelpad = 0)
    ax2.bar(x_axis, y_axis, align = 'center', color='r')
    plt.title("Popularity of Genres",bbox={'facecolor':'k', 'pad':5},color='w',fontsize = 30)
    plt.show()
```

```python
#merging two dataframes movies and ratings
movie_data_ratings_data=base.merge(movies_data,on = 'item_id',how = 'inner')
#creating high rated and low rate movies dataset
high_rated = movie_data_ratings_data[movie_data_ratings_data['rating']>4.0]
low_rated = movie_data_ratings_data[movie_data_ratings_data['rating']<2.0]

print('Popularity of Genres for high rated movies over the entire population:')
DisplayGenreCount(high_rated)
print('Popularity of Genres for low rated movies over the entire population:')
DisplayGenreCount(low_rated)
```

```python
merged_data = mean_rating_per_item_id.merge(data,on='item_id' ,how ='inner')
merged_data.columns = ['item_id','average rating per movie', 'rating']
merged_data['gap_rating'] = merged_data['rating'] - merged_data['average rating per movie']
merged_data = merged_data.drop(['rating'], axis = 1)


groupby_movie_id = merged_data.groupby(['item_id'])
groupby_movie_id_mean = groupby_movie_id.mean().reset_index(['item_id'])


test = pd.read_csv('/Users/adiozeri/u1.test',delimiter='\t',names=data_columns)
mean_rating = groupby_movie_id_mean['average rating per movie'].mean()

for i, row in test.iterrows():
    row_from_base = groupby_movie_id_mean[groupby_movie_id_mean['item_id'] == row['item_id']]
    if row_from_base.empty:
        test.at[i,'predict'] = mean_rating
    else:
        test.at[i,'predict'] = (row_from_base.iloc[0]['average rating per movie'] + row_from_base.iloc[0]['gap_rating'])

test['error'] = (test['predict']-test['rating']).abs()
test_mae = test['error'].mean()

print('Mean Absolut Error of the average rating per movie model:')
display(test_mae)
```

```python
data_by_item_id = data.groupby(['item_id'])
mean_rating_per_item_id = data_by_item_id.mean().reset_index('item_id').sort_values('rating', ascending=False)

columns_movies_recommandation = ['user_id','item_id']
movies_to_recommand_average = pd.DataFrame(columns = columns_movies_recommandation)
train_user_id_movie_id = base.drop(['rating', 'timestamp'], axis = 1)
test_unique_users = (test['user_id']).unique()
for user_id in test_unique_users:
    rows_with_user_id = train_user_id_movie_id[train_user_id_movie_id['user_id'] == user_id]
    movies_wasnt_watched = mean_rating_per_item_id[~mean_rating_per_item_id.item_id.isin(rows_with_user_id.item_id)]
    top_20_movies_wasnt_watched = movies_wasnt_watched.head(20).drop(['rating'],axis=1)
    movies_to_recommand_average = movies_to_recommand_average.append(top_20_movies_wasnt_watched.assign(user_id = user_id))
print('top 20 movies with the best average rating for each user that this user havent seen already')
display(movies_to_recommand_average)
```

```python
movies_to_recommand_random = pd.DataFrame(columns = columns_movies_recommandation)
for user_id in test_unique_users:
    rows_with_user_id = train_user_id_movie_id[train_user_id_movie_id['user_id'] == user_id]
    movies_wasnt_watched = mean_rating_per_item_id[~mean_rating_per_item_id.item_id.isin(rows_with_user_id.item_id)]
    top_20_movies_wasnt_watched = movies_wasnt_watched.sample(20).drop(['rating'],axis=1)
    movies_to_recommand_random = movies_to_recommand_random.append(top_20_movies_wasnt_watched.assign(user_id = user_id))
print('random 20 movies to recommand for each user which the user havent seen already:')
display(movies_to_recommand_random)
```

```python
def displayPrecisionAndRecallForAverageAndRnadom(users, test, recommandation_average, recommandation_random ):
    precision_df = pd.DataFrame(columns=['user_id','precision_average', 'precision_random'])
    recall_df = pd.DataFrame(columns=['user_id','recall_average', 'recall_random'])
    for user_id in users:
        rows_with_user_id_from_test = test[test['user_id'] == user_id]
        rows_with_user_id_from_recommandation_average = recommandation_average[recommandation_average['user_id'] == user_id]
        rows_with_user_id_from_recommandation_random = recommandation_random[recommandation_random['user_id'] == user_id]
        rows_with_user_id_and_recommand_item_id_average_length = len(rows_with_user_id_from_test.merge(rows_with_user_id_from_recommandation_average, on=['item_id'], how='inner'))
        rows_with_user_id_and_recommand_item_id_random_length = len(rows_with_user_id_from_test.merge(rows_with_user_id_from_recommandation_random, on=['item_id'], how='inner'))

        precision_averge = rows_with_user_id_and_recommand_item_id_average_length/20
        recall_average = rows_with_user_id_and_recommand_item_id_average_length/len(rows_with_user_id_from_test)
        print(precision_averge)
        print(recall_average)
        
        precision_random = rows_with_user_id_and_recommand_item_id_random_length/20
        recall_random = rows_with_user_id_and_recommand_item_id_random_length/len(rows_with_user_id_from_test)

        precision_df = precision_df.append({'user_id' : user_id, 'precision_average': precision_averge, 'precision_random': precision_random},ignore_index=True)
        recall_df = recall_df.append({'user_id' : user_id,'recall_average': recall_average, 'recall_random': recall_random},ignore_index=True)

    print('Average precision for the average rating model:')
    display(precision_df['precision_average'].mean())
    print('Average precision for the random model:')
    display(precision_df['precision_random'].mean())
    print('Average recall for the average rating model:')
    display(recall_df['recall_average'].mean())
    print('Average recall for the random model:')
    display(recall_df['recall_random'].mean())
```

```python
displayPrecisionAndRecallForAverageAndRnadom(test_unique_users, test, movies_to_recommand_average, movies_to_recommand_random)
print('The average model gives as better precision and recall results')
print('The reason is that the average model considers the average rating of movies, and was calculated with movies with high rating')
```
